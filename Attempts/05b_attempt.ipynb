```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gaussian_noise(x, mean, std):
    noise = np.random.normal(mean, std, size = x.shape)
    x_noisy = x + noise
    return x_noisy

mean = 0
std = 0.2

#create dataset
x = np.arange(-3,3, 0.01, dtype=float)
y = np.sin(x)
x = gaussian_noise(x, mean, std)


plt.rcParams["figure.figsize"] = (5,5)
plt.scatter(x,y)


```
